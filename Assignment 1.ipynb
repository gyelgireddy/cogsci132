```python
import numpy as np
import matplotlib.pyplot as plt
```

# Question 1

```python
# x = np.array(range(1, 11))
x = np.arange(-5, 5, 0.01)
y = np.sin([2**i for i in x])
```

```python
plt.plot(x,y)
plt.xlabel("x", fontsize = 10)
plt.ylabel ("sin(2^x)", fontsize = 10)
plt.title("Question 1")
plt.xticks(range(-5, 5))
plt.show()
```

# Question 2

```python
x = np.arange(-5, 5, 0.1)
y = [np.sin(2**i)/np.sin(2**(i + 0.1)) for i in x]
plt.plot(x,y)
plt.xlabel("x", fontsize = 10)
plt.ylabel("g", fontsize = 10)
plt.title("Question 2")
plt.xticks(range(-5, 5))
plt.show()
```

# Question 3a

```python
def f(n):
    res = 0
    for i in range(1, n+1):
        res = res + 1/(i+1)
    return res
```

# Question 3b

```python
a = np.arange(1, 101)
b = [f(i) for i in a]
plt.plot(a, b)
plt.xlabel ("a", fontsize = 10)
plt.ylabel ("f(a)", fontsize = 10)
plt.title("Question 3b")
plt.show()
```

# Question 4

```python
x = np.random.normal(0,1,10000) 
y = np.sin(x)
plt.hist(y, 20)
plt.xlabel("x-standard normal",  fontsize = 10)
plt.ylabel("count", fontsize=10)
plt.title("Question 4")
plt.show()
```

# Question 5

```python
x = np.random.uniform(0, 1.5, 10000)
y = [np.exp(i) for i in x]
plt.hist(y, 20)
plt.xlabel("x-standard uniform",  fontsize = 10)
plt.ylabel("count", fontsize=10)
plt.title("Question 5")
plt.show()
```

# Question 6a

```python
x=np.random.normal(0,1,1000)
y= np.sort(x)
diff = []
for i in range(0, len(x)-1):
    diff = np.append(diff, y[i+1] - y[i])

plt.hist(diff, 20)
plt.xlabel("x-standard normal",  fontsize = 10)
plt.ylabel("count", fontsize=10)
plt.title("Question 6a")
plt.show()
```
