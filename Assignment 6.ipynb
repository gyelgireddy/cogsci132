```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

# Question 1

```python
def contains(n=10000):
    start = np.random.uniform(-10, 10, n);
    end = np.random.uniform(-10, 10, n);
    regions = list(zip(start, end))
    return regions
```

# Question 2

```python
regions = contains()
probs = []
count = 0
total_prob = sum([1/(region[1]-region[0]) for region in regions])
for region in regions:
    if region[1] >=1:
        count += 1/(region[1]-region[0])
probs.append(count/total_prob)
print(probs)
```

# Question 3

```python
probs=[]
x=np.arange(-10,10,0.01)
total_prob = sum([1/(region[1]-region[0]) for region in regions])
for i in x: 
    count=0
    for region in regions:
        if region[0] <= i and i <= region[1]:
            count += 1/(region[1]-region[0])
    probs.append(count/total_prob)
plt.xlabel("x")
plt.ylabel("Probabilities")
plt.plot(x,probs)
plt.show()
```

This function looks like negative exponential function because as we get farther from zero the probabibility decreases.

# Question 6

```python
num_regions = [10,100,1000]
fig, axes = plt.subplots(1, 3)
for number in num_regions:
    regions = contains(number) 
    probs=[] 
    x=np.arange(-10,10,0.01) 
    total_prob=sum([1/(region[1]-region[0]) for region in regions]) 
    for i in x: 
        count=0
        for region in regions:
            if region[0]<=i and i<=region[1]:
                count+=1/(region[1]-region[0])
        probs.append(count/total_prob)
    axes[num_regions.index(number)].plot(x,probs)
plt.show()
```

As the number of consequential regions increases, it looks more like a exponential graph.
