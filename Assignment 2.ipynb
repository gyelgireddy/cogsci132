# Question 1a

Rescorla-Wagner Model: $V_{light}$ = $α_{light}$β(1-$V_{light}$)<br> 
For initial association: 0.05 <br>
$α_{light}$ = 0.5 β = 0.1


```python
x = [i for i in range(20)]
y = [0] * 20
z = [0] * 20
y[0] = 0.05
z[0] = 0.5
for i in range(0, 19):
    y[i+1] = y[i] + (0.5*0.1*(1-y[i]))
    z[i+1] = z[i] + (0.5*0.1*(1-z[i]))
plt.plot(x, y,linewidth =1)
plt.plot(x, z,linewidth =1)
plt.xlabel("Number of pairings", fontsize = 10)
plt.ylabel ("Association of Strength", fontsize = 10)
plt.title("Question 1a")
plt.show()
```

# Question 2

association for light and food is 0.8 <br>
$α_{light}$ = 0.2<br>
$V_{light}$ = $V_{bell}$ + 0.8 <br>
$V_{bell}$ = (0.5) (0.1) (0.2 - 2 $V_{bell}$)

```python
x = [0] * 75
y = [0] * 75
x[0] = 0.8
y[0] = 0

for i in range(0, 74):
    x[i+1] = x[i] + (0.5*0.1*(1-x[i]-y[i]))
    y[i+1] = y[i] + (0.5*0.1*(1-x[i]-y[i]))
    
plt.plot(x, linewidth =1, label = "association strength between light and food")
plt.plot(y, linewidth =1, label = "association strength between bell and food")
plt.xlabel("Number of pairings", fontsize = 10)
plt.ylabel ("Association of Strength", fontsize = 10)
plt.title("Question 2")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

# Question 3a

alternate trials

```python
x = [0] * 101
x[0] = 0
for i in range(0, 50):
    x[2*i+1] = x[2*i] + (0.5*0.1)*(1- x[2*i])
    x[2*i+2] = x[2*i+1] - (0.5*0.1)*(x[2*i+1])
plt.plot(x, linewidth =1)
plt.xlabel("Number of pairings", fontsize = 10)
plt.ylabel ("Association of Strength", fontsize = 10)
plt.title("Question 3a")
plt.show()
```

The association strength is 0.5. This happens because half of the time its exposed to light and the other half it isn't so the association isn't as strong. When you eventually go to the infinity, the association strength will go to 0.5.

# Question 3b

Probability P = 0.5 Trials = 50

```python
x = [0] * 50
x[0] = 0 #what should initial be??
p = 0.5
for i in range(0, 49):
    if (np.random.random()> p):
        x[i+1] = x[i] + (0.5*0.1*(1-x[i]))
    else:
        x[i+1] = x[i] - (0.5*0.1*(x[i]))

plt.plot(x, linewidth =1)
plt.xlabel("Number of pairings", fontsize = 10)
plt.ylabel ("Association of Strength", fontsize = 10)
plt.title("Question 3b")
plt.show()
```

Probability P = 0.5 Trials = 1000

```python
x = [0] * 10000
x[0] = 0 #what should initial be??
p = 0.5
for i in range(0, 9999):
    if (np.random.random()> p):
        x[i+1] = x[i] + (0.5*0.1*(1-x[i]))
    else:
        x[i+1] = x[i] - (0.5*0.1*(x[i]))
plt.plot(x, linewidth =1)
plt.xlabel("Number of pairings", fontsize = 10)
plt.ylabel ("Association of Strength", fontsize = 10)
plt.title("Question 3b")
plt.show()
```

As probability decreases, association strength decreases. As the subject is exposed more to the stimuli, the association increases.  Associative strength depends on the sum.
